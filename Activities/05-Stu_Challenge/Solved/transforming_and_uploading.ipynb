```python
import os
import io
import boto3
import pandas as pd
```

```python
# Set up s3 resource and Object instance
s3 = boto3.resource("s3")
bucket = "trilogy-student-12345"
key = "pokemon-data/poke_skills.csv"
obj = s3.Object(bucket, key)
```

```python
# Create path to file and instantiate string buffer
path = os.path.join("data", "pokemon.csv")
buffer = io.StringIO()
```

```python
# Load iris data
df = pd.read_csv(path)
df.head()
```

```python
# Create new field that determines if pokemon has above average strength
df["above avg. strength"] = df["Total"].map(
    lambda row: True if row >= df["Total"].mean() else False
    )
df.head()
```

```python
# Save csv to buffer and upload to s3
df.to_csv(buffer, index=False, header=True)
res = obj.put(Body=buffer.getvalue())
```

```python
# BONUS: Message for checking upload status
status_code = res["ResponseMetadata"]["HTTPStatusCode"]
if status_code == 200:
    print(f"File uploaded successfully to {bucket}/{key}")
else:
    print(f"Error uploading file. Returned status code: {status_code}")
```
